## Exploded View

A central "Paper" node with the other components revolving.

```python
#| label: exploded-view
import matplotlib.pyplot as plt

def draw_exploded_view():
    fig, ax = plt.subplots(figsize=(12, 10))
    ax.set_facecolor('white')

    # Coordinates
    center = (0.5, 0.5)

    # Outlying Nodes (x, y, label, color)
    nodes = [
        (0.5, 0.85, "DESIGN\n(Protocol, Hypotheses)", "#66c2a5"), # Top
        (0.5, 0.15, "ETHICS\n(IRB, Consent)", "#fc8d62"),       # Bottom
        (0.15, 0.5, "DATA\n(Raw Data, Materials)", "#8da0cb"),  # Left
        (0.85, 0.5, "CODE & ANALYSIS\n(Scripts, Models)", "#e78ac3"), # Right
        (0.2, 0.8, "FUNDING\n(Grants)", "#a6d854"),             # Top Left Corner
        (0.8, 0.2, "METADATA\n(DOIs, Attribution)", "#ffd92f")  # Bottom Right Corner
    ]

    # Draw Connections first
    for x, y, label, color in nodes:
        ax.plot([center[0], x], [center[1], y], color='grey', linestyle='--', linewidth=1, zorder=1)

    # Draw Center Node (The Paper)
    ax.add_patch(plt.Rectangle((0.4, 0.45), 0.2, 0.1, color='lightgrey', alpha=0.5, ec='black', ls='--'))
    ax.text(0.5, 0.5, "The PDF\n(Container)", ha='center', va='center', fontsize=12, fontweight='bold', color='grey')

    # Draw Outlying Nodes (use larger box so two-line labels fit)
    box_w, box_h = 0.20, 0.14
    # Light colors: use black text for contrast
    light_hex = {'#ffd92f', '#a6d854', '#66c2a5'}
    for x, y, label, color in nodes:
        # Draw Box (transform=ax.transData so coordinates match axes)
        rect = plt.Rectangle((x - box_w/2, y - box_h/2), box_w, box_h, color=color, ec='black', lw=2, zorder=10, transform=ax.transData)
        ax.add_patch(rect)
        # Draw Label: black text on light backgrounds, white on dark
        text_color = 'black' if color.lower() in light_hex else 'white'
        ax.text(x, y, label, ha='center', va='center', fontsize=9, fontweight='bold', color=text_color, zorder=11)

    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.axis('off')
    ax.set_title("Exploded View: Research Components", fontsize=18, fontweight='bold')

    plt.tight_layout()
    plt.show()

# Run the function
draw_exploded_view()
```

## The Science Iceberg

Draws a schematic iceberg, visualizing the specific components from your summary that sit above vs. below the "waterline."

```python
#| label: iceberg
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_iceberg():
    fig, ax = plt.subplots(figsize=(10, 12))

    # --- Backgrounds ---
    # Sky
    ax.add_patch(patches.Rectangle((0, 0.7), 1, 0.3, color='#CCEEFF', zorder=0))
    # Water
    ax.add_patch(patches.Rectangle((0, 0), 1, 0.7, color='#004488', zorder=0))

    # --- The Iceberg Shape ---
    # Polygon coordinates (x, y) relative to 0-1 scale
    iceberg_x = [0.4, 0.5, 0.6, 0.75, 0.65, 0.5, 0.35, 0.25]
    iceberg_y = [0.9, 1.0, 0.85, 0.4, 0.1, 0.05, 0.1, 0.4]

    iceberg = plt.Polygon(list(zip(iceberg_x, iceberg_y)), color='white', alpha=0.9, ec='#ADD8E6', lw=2, zorder=1)
    ax.add_patch(iceberg)

    # --- The Water Line ---
    ax.axhline(y=0.7, color='white', linestyle='--', linewidth=2, alpha=0.7)
    ax.text(0.02, 0.71, "Surface (The PDF)", color='white', fontweight='bold', fontsize=12)
    ax.text(0.02, 0.68, "Deep Review (The Evidence)", color='white', fontweight='bold', fontsize=12)

    # --- Text Labels (The Components) ---

    # Zone 1: The Tip (Visible)
    ax.text(0.5, 0.85, "The Manuscript\nFigures & Tables", ha='center', fontsize=11, fontweight='bold', color='#003366')
    ax.text(0.5, 0.78, "Lay Summaries", ha='center', fontsize=10, color='#003366')

    # Zone 2: Just Below Surface
    ax.text(0.5, 0.60, "Metadata & Ethics\nStatements", ha='center', fontsize=11, fontweight='bold', color='#003366')

    # Zone 3: Middle Depth
    ax.text(0.5, 0.45, "Raw Data & Datasets\nAnalysis Code", ha='center', fontsize=11, fontweight='bold', color='#003366')

    # Zone 4: Deep Foundation
    ax.text(0.5, 0.30, "Protocols & Preregistration\nGrant Proposals", ha='center', fontsize=11, fontweight='bold', color='#003366')

    # Zone 5: The Bottom
    ax.text(0.5, 0.15, "Emerging Artifacts\n(Performative / Time-based)", ha='center', fontsize=10, color='#003366')

    # --- Formatting ---
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.axis('off')
    ax.set_title("The Unbundled Iceberg", fontsize=18, fontweight='bold', color='#333333')

    plt.tight_layout()
    plt.show()

# Run the function
draw_iceberg()
```
